# Pandas Exercises

```
import pandas as pd

url = "https://raw.githubusercontent.com/MedlyticsUniversal/Data/main/Week1/diabetes.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
```

Read in csv above to a Pandas DataFrame named `data`.

```
# YOUR CODE HERE
data = pd.read_csv(url, names=names) # read in csv
```

Print out the age and class of the 100th patient.

```
# YOUR CODE HERE
data.iloc[99, 7:9] #dice 100th row with age and class columns only
```

Write code to figure out how many people are classified as not having diabetes (class = 0).

```
# YOUR CODE HERE
num_wo_diabetes = data[data["class"] == 0].count() # count number of people without diabetes
print(num_wo_diabetes) # print number of people without diabetes
```

Write code to figure out how many people are classified as having diabetes (class = 1).

```
# YOUR CODE HERE
num_with_diabetes = data[data["class"] == 1].count() # count the number of people who have diabetes
print(num_with_diabetes) # print number of people with diabetes
```

What percent of patients have diabetes in this study?

```
# YOUR CODE HERE
percent_diabetic = (num_with_diabetes / (num_with_diabetes + num_wo_diabetes)) * 100 # calc percent of diabetics
print(percent_diabetic) # print out the percent of diabetics
```

Print out the `"pres"` and `"class"` (no other columns) for the first 5 people.

```
# YOUR CODE HERE
data.loc[0:4, ['pres','class']] # dice the first five rows with just pres and class columns
```

Run the cell below. It will print out the correlation among the columns.

```
data.corr()
```

What column has the highest correlation with class? Find the mean() value of that column.

```
# YOUR CODE HERE
data["class"].mean() #calculate mean of class column
```

Create a new DataFrame with only people that are above the mean found in the above cell.

```
# YOUR CODE HERE
data_above_mean = data["class"] > data["class"].mean() # create new dataframe with people above class mean
data_above_mean
```

Create a new DataFrame that only includes people who are above the average value for each column in the dataset. In other words, find the average value for each column, then find which users have values above all of these average column values.

```
# YOUR CODE HERE
data_copy = data.copy() # duplicate dataframe

for column in data.columns: #iterate through each column and remove users with values below all average column values
    data_copy = data_copy[data_copy[column] > data[column].mean()]

data_copy
```

Ignoring the index (resetting the index to 0) and using the DataFrame from the cell above, sort the values based on `"skin"`, `"test"`, and `"pedi"` values in ascending order. Please write this in one line.

```
# YOUR CODE HERE
data_copy = data_copy.reset_index(drop=True) # reset index
data_copy.sort_values(by=["skin", "test", "pedi"], ascending=True) # sort by skin, test, pedi in ascengin order
```

Now, let's find the 4 largest values for `"age"`.

```
# YOUR CODE HERE
data_copy.nlargest(4, "age") # get the largest 4 values for age
```

Find the smallest 4 values for `"pedi"` and `"plas"` in reverse order (smallest to largest).

```
# YOUR CODE HERE
data_copy.nsmallest(4, "pedi") # smallest 4 values for pedi
data_copy.nsmallest(4, "plas") # smallest 4 values for plas
```
